```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Python Projects/Classifiers for supervised learning tasks/logs.csv')
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X = df[["Source Port", "Destination Port", "NAT Source Port", "NAT Destination Port","Bytes","Bytes Sent","Bytes Received","Packets","Elapsed Time (sec)","pkts_sent","pkts_received"]]
Y = df["Action"]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from xgboost import XGBClassifier


#inicialize XGBoost
model = XGBClassifier(max_depth=50, min_child_weight=1,  n_estimators=200,n_jobs=-1 , verbose=1,learning_rate=0.16)

# Add silent=True to avoid printing out updates with each cycle
model.fit(X_train, Y_train)

#make prediction
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(Y_test, y_pred))
print(classification_report(Y_test, y_pred))
```
