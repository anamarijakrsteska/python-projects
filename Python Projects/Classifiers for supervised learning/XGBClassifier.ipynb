# **Digital forensics**

```python
#Importing libraries

import pandas as pd

from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
df = pd.read_csv('logs.csv')
```

```python
df.head()
```

```python
df
```

```python
X = df[["Source Port", "Destination Port", "NAT Source Port", "NAT Destination Port","Bytes","Bytes Sent","Bytes Received","Packets","Elapsed Time (sec)","pkts_sent","pkts_received"]]
Y = df["Action"]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0, stratify=Y)
```

```python
#inicializing XGBoost
model = XGBClassifier(max_depth=5, min_child_weight=1, n_estimators=200, n_jobs=-1, learning_rate=0.15)

# Adding silent=True to avoid printing out updates with each cycle
model.fit(X_train, Y_train)

#making prediction
y_pred = model.predict(X_test)
```

```python
print(confusion_matrix(Y_test, y_pred))
print(classification_report(Y_test, y_pred))
```

```python
accuracy = accuracy_score(Y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
print("Train accuracy:", model.score(X_train, Y_train))
print("Test accuracy:", model.score(X_test, Y_test))
```

Train and test are: almost identical and both extremely high
- This is good generalization
- Model is not overfitting
